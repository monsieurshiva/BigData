### Query Operators

```python
db.emp.find({email: {$eq:"ericcarter176@slingacademy.com"}}) // equal
```

image.png

```python
db.emp.find({email: {$ne:"ericcarter176@slingacademy.com"}}) // not equal
```

image.png

```python
db.emp.find({age:{$in :[20,22,28,45]}})
```

image.png

```python
db.emp.find({$and :[{email: {$eq:"ericcarter176@slingacademy.com"}}, {age:{$in :[20,22,28,45]}}]})
```
