**JOINS**

```python
--Inner Join
select e.name,e.age,s.salary from employee e join salary s on (s.id=e.id)
```

image.png

```python
--Left Outer Join
select e.name,e.age,s.salary from employee e left outer join salary s on (s.id=e.id)
```

```python
--Right Outer Join
select e.name,e.age,s.salary from employee e right outer join salary s on (s.id=e.id)
```

**Views**

```python
create view if not exists earningemployee as select e.name,e.age,s.salary from employee e right outer join salary s on (s.id=e.id) where s.salary>0; 
```
